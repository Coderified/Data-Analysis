```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X
```

```python
X.shape
```

```python
y
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
ss=StandardScaler()

Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.25,random_state=42)
Xtrainsc = ss.fit_transform(Xtrain)
Xtestsc = ss.transform(Xtest)
```

```python
from sklearn.cluster import KMeans
```

```python

wcss = []

for k in range(1,11):
    km = KMeans(n_clusters = k, init='k-means++')
    km.fit(Xtrainsc)
    wcss.append(km.inertia_)

    
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,12))
plt.show()
```

```python
km = KMeans(n_clusters = 3, init='k-means++')
```

```python
km.fit(Xtrainsc)
```

```python
ytrainpreds = km.predict(Xtrainsc)
```

```python
ytestpreds = km.predict(Xtestsc)
```

```python
from sklearn.metrics import silhouette_score 
```

```python
sil_coeff=[]
for i in range(2,12):
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(Xtrainsc)
    score = silhouette_score(Xtrainsc,km.labels_)
    sil_coeff.append(score)
```

```python
sil_coeff
```

```python
plt.plot(range(2,12),sil_coeff)
plt.xticks(range(2,11))
plt.show()
```
