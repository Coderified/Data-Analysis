# https://keras.io/api/applications/resnet/

```python
from tensorflow.keras.applications.resnet50 import ResNet50

from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions

from tensorflow.keras.preprocessing import image

from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img

from tensorflow.keras.models import Sequential

import numpy as np

from glob import glob

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img_size=[224,224]

train_path='E:\oldBackups\Projects & Courses\Datasets\Car PredictionDatasets\Train'

valid_path='E:\oldBackups\Projects & Courses\Datasets\Car PredictionDatasets\Test'
```

```python
resnet = ResNet50(input_shape=[224,224]+[3],weights='imagenet',include_top=False)
```

```python
resnet.summary()
```

```python
# We are yet to add the op layer in resnet
```

```python
# Weights of neurons in layers will not be altered
for layer in resnet.layers:
    layer.trainable=False
```

```python
folders=glob('E:\oldBackups\Projects & Courses\Datasets\Car PredictionDatasets\Train\*')
# Asterix at the end is important else it will just direct to the main folder link
```

```python
folders
```

```python
x=Flatten()(resnet.output)
```

```python
prediction=Dense(len(folders),activation='softmax')(x)


#Model Creation
model=Model(inputs=resnet.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### For better model ,data augmentation is done

### Here images are rotated, zoomed in, flipped and such oprns carried out to bettter prepare the model for predicition

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1.0/255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True
                                  )

test_datagen=ImageDataGenerator(rescale=1.0/255)
```

```python
training_set=train_datagen.flow_from_directory('E:\oldBackups\Projects & Courses\Datasets\Car PredictionDatasets\Train',
                                               target_size = (224, 224),
                                               batch_size = 32,
                                               class_mode = 'categorical')
```

```python
test_set=test_datagen.flow_from_directory('E:\oldBackups\Projects & Courses\Datasets\Car PredictionDatasets\Test',
                                         target_size=(224,224),
                                         batch_size=32,
                                         class_mode='categorical')
```

```python
# fit the model
# Run the cell. It will take some time to execute

r = model.fit_generator(
    training_set,
    validation_data=test_set,
    epochs=40,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
   )
```

```python
# Accuracy and Losses comparison 

# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
