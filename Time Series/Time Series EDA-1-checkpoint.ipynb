```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_tsla = pdr.get_data_yahoo('TSLA')
df_tsla.info()
```

```python
df_tsla.head()
```

```python
df_tsla['Low'].plot(figsize=(12,4),xlim=['2019-01-01','2022-05-05'],c='Red',ls='--')
```

```python
# To get data in bet specific time frame
# ":" is used here instead of "," as in plot


index=df_tsla.loc['2019-01-01':'2022-05-05'].index
open_price = df_tsla.loc['2019-01-01':'2022-05-05']['Open']
```

```python
open_price.head()
```

```python
figure,axis=plt.subplots(figsize=(12,12))
#autofmt formats the graph to adjust x axis to make it legible
figure.autofmt_xdate()

axis.plot(index,open_price)
```

```python
# Date_time is working as index
# to make it a clm, use reset_index
df_tsla=df_tsla.reset_index()
df_tsla.info()
```

```python
# To make date time as index

# USe set_index, drop=True
# if drop=False, old clm will remain as such

df_tsla=df_tsla.set_index('Date',drop=True)
df_tsla.head()
```


### https://towardsdatascience.com/using-the-pandas-resample-function-a231144194c4

## Article on Time Series Resampling

```python

df_tsla.resample(rule='SMS').mean() #--sEMI MONTHLY MEAN
df_tsla.resample(rule='Q').max() # -- QUARRTERLY MAX
df_tsla.resample(rule='M').min() # -- MONTHLY MIN
```

```python
df_tsla.resample(rule='W').max()['Open'].plot(figsize=(10,10))
```

```python
df_tsla.resample(rule='M').max()['Volume'].plot(figsize=(10,10))
```

```python
df_tsla.resample(rule='Q').max()['High'].plot(figsize=(10,10))
```

```python
df_tsla.resample(rule='Y').max()['Close'].plot(figsize=(10,10))
```

# Rolling for smoothening

```python
df_tsla['10 days Rolling - Open'] = df_tsla['Open'].rolling(20).mean()
```

```python
df_tsla[['Open','10 days Rolling - Open']].plot(figsize=(15,10))
```
