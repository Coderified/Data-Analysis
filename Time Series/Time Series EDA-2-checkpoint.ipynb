```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_tsla = pdr.get_data_yahoo('TSLA')
df_tsla.head(-5)
```

## MA(Rolling) vs WMA vs #EMA

####  https://www.investopedia.com/ask/answers/071414/whats-difference-between-moving-average-and-weighted-moving-average.asp#:~:text=The%20main%20difference%20between%20simple,more%20weight%20to%20current%20data.

## Moving Average

```python
df_tsla['10 days Rolling - Open'] = df_tsla['Open'].rolling(10,min_periods=2).mean()
```

```python
df_tsla[['Open','10 days Rolling - Open']].plot(figsize=(20,10),xlim=['2020-1-5','2021-5-5'])
```

## Expanding - Cumulative Moving Avg

```python
df_tsla['Expanding']=df_tsla['Open'].expanding().mean()
```

```python
df_tsla.head()
```

## Ewm - Expo. Weighted Moving Avg

```python
df_tsla['ewma_alpha=0.1']=df_tsla['Open'].ewm(alpha=0.1,adjust=True).mean()
df_tsla['ewma_span=3']=df_tsla['Open'].ewm(span=3,adjust=True).mean()
df_tsla.drop('ewma_span=1',inplace=True,axis=1)
```
