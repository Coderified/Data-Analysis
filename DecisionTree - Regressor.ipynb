```python
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
df = pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
```

```python
df.head()
```

```python
tar = dataset['target']
```

```python
import seaborn as sns
```

```python
sns.boxplot(df)
plt.show()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df,tar,test_size=0.3,random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

```python
plt.scatter(x=y_test,y=preds,color='blue', label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Fit')
plt.show()
```

```python
sns.histplot((y_test-preds),kde=True)
plt.show()
```

```python
plt.scatter(x=y_test,y=y_test-preds,color='blue', label='residuals')
plt.axhline(y=0, color='green', linestyle='--')

plt.show()
```

```python
params={
    'criterion' : ["squared_error", "friedman_mse", "absolute_error","poisson"],
    'splitter' : ["best", "random"],
    'max_depth':[1,2,3,4,5,10,15],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator = dt, param_grid=params,scoring='neg_mean_squared_error',cv=6)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_index_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
plt.scatter(x=y_test,y=y_pred,color='blue', label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Perfect Fit')
plt.show()
```

```python
sns.histplot((y_test-y_pred),kde=True)
plt.show()
```

```python
plt.scatter(x=y_test,y=y_test-y_pred,color='blue', label='residuals')
plt.axhline(y=0, color='green', linestyle='--')

plt.show()
```

```python
print(r2_score(y_test,preds))
print(mean_absolute_error(y_test,preds))
print(mean_squared_error(y_test,preds))
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
final_model = DecisionTreeRegressor(criterion= 'squared_error',
 max_depth= 4,
 max_features = 'sqrt',
splitter='best')
```

```python
final_model.fit(X_train,y_train)
```

```python
##Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(final_model,filled=True)
plt.show()
```
